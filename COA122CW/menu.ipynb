```python
from IPython.display import clear_output
import ipywidgets as widgets

import gameSearch
import gameRent
import gameReturn
import InventoryPruning

import pandas as pd

'''
Student ID: F328307
'''

# Submit button code

def on_search_clicked(b):
    with output:
        clear_output()
        option = options.value
        if option == "Title":
            title = search_name_input.value
            data = gameSearch.searchByTitle(title)
        elif option == "Genre":
            genre = search_name_input.value
            data = gameSearch.searchByGenre(genre)
            
        elif option == "Platform":
            platform = search_name_input.value
            data = gameSearch.searchByPlatform(platform)
            
        display(pd.DataFrame.from_records(data))
            
def on_rent_clicked(b):
    with output:
        message, newGame = gameRent.rentGame(customerID_input.value, gameIDrent_input.value)
        print(message)
        display(pd.DataFrame(newGame, index=['Game']))
        
        
def on_return_clicked(b):
    with output:
        message = gameReturn.returnGame(gameIDreturn_input.value, rating_input.value, comments_input.value)
        print(message)
        gameIDreturn_input.value = ""
        rating_input.value = ""
        comments_input.value = ""
        

# Click Events for menu buttons

def on_gameSearch_clicked(b):
    with output:
        clear_output()
        display(options, search_name_input, btn_search)
        btn_search.on_click(on_search_clicked)

    
def on_gameRent_clicked(b):
    with output:
        clear_output()
        display(customerID_input, gameIDrent_input, btn_rent)
        btn_rent.on_click(on_rent_clicked)

        
def on_gameReturn_clicked(b):
    with output:
        clear_output()
        display(gameIDreturn_input, rating_input, comments_input, btn_return)
        btn_return.on_click(on_return_clicked)

        
def on_inventoryPruning_clicked(b):
    with output:
        clear_output()
        InventoryPruning.ratingsBarChart()
        gamesToRemove = InventoryPruning.inventoryPruning()
        print("Recommended games to remove based on average ratings:", gamesToRemove)


# Creates buttons for each menu option
btn_gameSearch = widgets.Button(description="Game Search")
btn_gameRent = widgets.Button(description="Rent Game")
btn_gameReturn = widgets.Button(description="Return Game")
btn_inventoryPruning = widgets.Button(description="Inventory Pruning")


# Game Search features
options = widgets.Dropdown(
    options=['Title', 'Genre', 'Platform'],
    description='Search by:',
    disabled=False,
)
search_name_input = widgets.Text(description="Name:")
btn_search = widgets.Button(description="Search Game")

# Rent Game features
customerID_input = widgets.Text(description="Customer ID:")
gameIDrent_input = widgets.Text(description="Game ID:")
btn_rent = widgets.Button(description="Rent Game")


# Return Game features
gameIDreturn_input = widgets.Text(description="Game ID:")
rating_input = widgets.Text(description="Rating /5:")
comments_input = widgets.Text(description="Comments:")
btn_return = widgets.Button(description="Return Game")


# Creates an output widget
output = widgets.Output()

# Links menu buttons to their event handlers
btn_gameSearch.on_click(on_gameSearch_clicked)
btn_gameRent.on_click(on_gameRent_clicked)
btn_gameReturn.on_click(on_gameReturn_clicked)
btn_inventoryPruning.on_click(on_inventoryPruning_clicked)

# Groups buttons together in a horizontal box
buttons = widgets.HBox([btn_gameSearch, btn_gameRent, btn_gameReturn, btn_inventoryPruning])

# Displays buttons and the output area
display(buttons, output)
```
